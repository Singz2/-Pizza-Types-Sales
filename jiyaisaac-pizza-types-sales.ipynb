```python
import pandas as pd

# load CSV files
order = pd.read_csv(r"C:\Users\TOSHIBA\Documents\Pizza+Place+Sales\pizza_sales\orders.csv")
order_details = pd.read_csv(r"C:\Users\TOSHIBA\Documents\Pizza+Place+Sales\pizza_sales\order_details.csv")
pizza = pd.read_csv(r"C:\Users\TOSHIBA\Documents\Pizza+Place+Sales\pizza_sales\pizzas.csv")
pizza_types = pd.read_csv(r"C:\Users\TOSHIBA\Documents\Pizza+Place+Sales\pizza_sales\pizza_types.csv", encoding='cp1252')

# Merge the datasets
# Step 1: Merge Order with Order Details on 'order_id'
df = pd.merge(order, order_details, on='order_id', how='left')

# Step 2: Merge with Pizzas on 'pizza_id'
df = pd.merge(df, pizza, on='pizza_id', how='left')

# Step 3: Merge with Pizza Types on 'pizza_type_id'
df = pd.merge(df, pizza_types, on='pizza_type_id', how='left')

# Ensure date column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Verify the merged DataFrame
display(df)




```

### Total Revenue/Sales

```python
### Total Revenue/Sales
df['revenue'] = df['quantity'] * df['price'] # quantity by price is equal  to revenue
total_revenue = df["revenue"].sum() # total sumation of revenue

print(f"Total Revenue: ${total_revenue:,.2f}")

### Total Quantity Sold
total_quantity = df["quantity"].sum() # total sumationn of quantity
print(f"Total Quantity Sold: {total_quantity:,}")

### Total Orders
total_orders = df["order_id"].nunique() # count unique pizza_type_id values  
print(f"Total Orders: {total_orders:,}")

### Number of Pizza Types
num_pizza_types = pizza_types["pizza_type_id"].nunique() # count unique pizza_type_id values
print(f"Number of pizza types: {num_pizza_types}")

### Average price of Pizzas
avg_price = pizza["price"].mean() # calculate mean for the price column
print(f"Average Pizza price: ${avg_price:.2f}")
```

### Peak Hours of Sales

```python

# Extract hour from time (assuming time is in HH:MM:SS format)
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour
peak_hours = df.groupby('hour')['revenue'].sum().sort_values(ascending=False)
print("Peak Hours of Sales (by Revenue):")
print(peak_hours.head())
```

### Top 5 Best selling pizza

```python
# extract the month from date column and grouping by it to calculate total revenue

df['day_of_week'] = df['date'].dt.day_name()
sales_by_day = df.groupby('day_of_week')['revenue'].sum().sort_values(ascending=False)
print("Total Sales by Day of the Week:")
print(sales_by_day)

print(f"Day with Most Sales: {sales_by_day.idxmax()}")
```

### Sales By Month

```python
#Extract month from the date column and grouping by it to calculate total revenue

df['month'] = df['date'].dt.month_name()
sales_by_month = df.groupby('month')['revenue'].sum().sort_values(ascending=False)
print("Sales by Month:")
print(sales_by_month)
print(f"month with Most Sales: {sales_by_month.idxmax()}")
```

### Pizza Types: With Poor sales

```python
#identify pizza types with lowwest sales grouping by name and summing quantity then sorting in ascending order 

low_performers = df.groupby("name")['quantity'].sum().sort_values().head(5)
print("Lowwwest Performing Pizza Types:")
print(low_performers)
```
